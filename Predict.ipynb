```python
from sklearn.linear_model import LinearRegression
x=[[1],[2],[3],[4],[5]]
y=[30,40,50,75,90]

model=LinearRegression()
model.fit(x,y)

hours=float(input("Enter Hour: "))
p_marks=model.predict([[hours]])

print(f'{hours} Hours -->{p_marks}')
```

```python
from sklearn.linear_model import LinearRegression

x=[[50],[60],[70],[80],[90]]
y=[1,2,3,4,5]

model=LinearRegression()
model.fit(x,y)

hours=float(input("Enter Hours: "))
pre_num=model.predict([[hours]])[0]


grade_map={1:'E',2:'D',3:'C',4:'B',5:'A'}
pre_num_rounded=round(pre_num)
pre_grade=grade_map.get(pre_num_rounded,'UNKNOWN')

print(f'{hours} Hour=> {pre_grade}({pre_num})')
```
