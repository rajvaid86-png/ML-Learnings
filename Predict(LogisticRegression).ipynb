```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

data={
    'weight':[150,170,140,130,180,200,120,110,190,210],
    'size':[7,8,6,5,9,10,5,4,9,10],
    'texture':[0,0,0,0,1,1,0,0,1,1],
    'label':['apple','apple','apple','apple','orange','orange','banana','banana','orange','orange']

}

df=pd.DataFrame(data)
print("Dataset:",df,"\n")

x=df[['weight','size','texture']]
y=df['label']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)

knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)

weight=float(input("enter weight:"))
size=float(input("enter size:"))
texture=float(input("enter texture:"))

prediction=knn.predict([[weight,size,texture]])

print("Test Data:",weight,size,texture)
print("Predicted Fruit:",prediction)

accuracy=knn.score(x_test,y_test)
print("\nModel Accuracy:",accuracy,"\n")
```
